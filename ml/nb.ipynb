```python
import os
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Embedding, Dense, GlobalAveragePooling1D, Concatenate, Dropout, Subtract, Multiply


# === Parameters ===
embedding_dim = 64
num_players = X.shape[1] // 2  # assume half columns = team1, half = team2
num_classes = len(np.unique(y))  # infer number of classes from y

# === Prepare inputs ===
team1 = X[:, :num_players].astype(np.int32)
team2 = X[:, num_players:].astype(np.int32)
y = y.astype(np.int32)

# === Build model ===
team1_input = Input(shape=(num_players,), name='team1_input')
team2_input = Input(shape=(num_players,), name='team2_input')

player_embedding = Embedding(input_dim=vocab_size, output_dim=embedding_dim, name='player_embedding')

team1_emb = player_embedding(team1_input)
team2_emb = player_embedding(team2_input)

team1_repr = GlobalAveragePooling1D()(team1_emb)
team2_repr = GlobalAveragePooling1D()(team2_emb)

team_diff = Subtract()([team1_repr, team2_repr])
team_mult = Multiply()([team1_repr, team2_repr])

combined = Concatenate()([team1_repr, team2_repr, team_diff, team_mult])

x = Dense(128, activation='relu')(combined)
x = Dropout(0.3)(x)
x = Dense(64, activation='relu')(x)
x = Dropout(0.3)(x)
x = Dense(32, activation='relu')(x)
x = Dropout(0.3)(x)
x = Dense(16, activation='relu')(x)
output = Dense(num_classes, activation='softmax')(x)

model = Model(inputs=[team1_input, team2_input], outputs=output)
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# === Train model ===
model.fit([team1, team2], y, batch_size=32, epochs=100, validation_split=0.1)
```

```python
import tensorflow.keras.backend as K
from tensorflow.keras.layers import Dense, Softmax, Multiply, Lambda

# Attention pooling helper function
def attention_pooling(embeddings):
    # embeddings shape: (batch, num_players, embedding_dim)
    scores = Dense(1)(embeddings)            # shape (batch, num_players, 1)
    weights = Softmax(axis=1)(scores)        # softmax over players dimension
    weighted = Multiply()([embeddings, weights])  
    output = Lambda(lambda x: K.sum(x, axis=1))(weighted)  # weighted sum over players
    return output

# Interaction block for team embeddings
def interaction_block(embeddings):
    # Apply a dense layer with ReLU to each player embedding (shared weights)
    transformed = Dense(embedding_dim, activation='relu')(embeddings)
    # Apply attention pooling to aggregate player embeddings into team vector
    pooled = attention_pooling(transformed)
    return pooled


# === Your existing setup ===
# ... assume team1_input, team2_input, player_embedding, etc. are defined

team1_emb = player_embedding(team1_input)  # shape: (batch, num_players, embedding_dim)
team2_emb = player_embedding(team2_input)

# Apply interaction block (dense + attention pooling) to each team
team1_repr = interaction_block(team1_emb)
team2_repr = interaction_block(team2_emb)

team_diff = Subtract()([team1_repr, team2_repr])
team_mult = Multiply()([team1_repr, team2_repr])

combined = Concatenate()([team1_repr, team2_repr, team_diff, team_mult])

x = Dense(128, activation='relu')(combined)
x = Dropout(0.3)(x)
x = Dense(64, activation='relu')(x)
x = Dropout(0.3)(x)
x = Dense(32, activation='relu')(x)
x = Dropout(0.3)(x)
x = Dense(16, activation='relu')(x)
output = Dense(num_classes, activation='softmax')(x)

model = tf.keras.Model(inputs=[team1_input, team2_input], outputs=output)
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.fit([team1, team2], y, batch_size=32, epochs=100, validation_split=0.1)
```

```python
import numpy as np
import pandas as pd
import os

from sklearn.preprocessing import LabelEncoder
import tensorflow as tf
from tensorflow.keras.layers import Input, Embedding, Dense, Concatenate, Lambda
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam

encoder = LabelEncoder()
X_flat = X.flatten()
X_encoded = encoder.fit_transform(X_flat).reshape(X.shape)

vocab_size = X_encoded.max() + 1  # input_dim for embedding

# === Model parameters ===
embedding_dim = 32
input_length = X_encoded.shape[1]
num_classes = len(np.unique(y))

# === Define Keras model ===
input_layer = Input(shape=(input_length,))

embedding_layer = Embedding(input_dim=vocab_size, output_dim=embedding_dim)(input_layer)

# Split into teams (first 5 players and last 5 players)
team1 = Lambda(lambda x: x[:, :5, :])(embedding_layer)
team2 = Lambda(lambda x: x[:, 5:, :])(embedding_layer)

# Average over team members to get team representation
team1_vec = Lambda(lambda x: tf.reduce_mean(x, axis=1))(team1)
team2_vec = Lambda(lambda x: tf.reduce_mean(x, axis=1))(team2)

# Concatenate team vectors
x = Concatenate()([team1_vec, team2_vec])
x = Dense(64, activation='relu')(x)
x = Dense(32, activation='relu')(x)

# Output layer
if num_classes == 2:
    output = Dense(1, activation='sigmoid')(x)
    loss = 'binary_crossentropy'
else:
    output = Dense(num_classes, activation='softmax')(x)
    loss = 'sparse_categorical_crossentropy'

# Compile model
model = Model(inputs=input_layer, outputs=output)
optimizer = Adam(learning_rate=1e-5)
model.compile(optimizer=optimizer, loss=loss, metrics=['accuracy'])

# === Train model ===
model.fit(X_encoded, y, epochs=1000, batch_size=128, validation_split=0.05)

# === Extract embeddings after training ===
player_embeddings = model.get_layer(index=1).get_weights()[0]  # Shape: (vocab_size, embedding_dim)

# Optional: map embeddings back to player IDs
player_ids = encoder.inverse_transform(np.arange(vocab_size))
embedding_lookup = dict(zip(player_ids, player_embeddings))
```
