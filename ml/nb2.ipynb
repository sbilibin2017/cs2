```python
import os
import pandas as pd
import numpy as np

PATH_TO_DATASET = "../data/datasets"
filename = os.listdir(PATH_TO_DATASET)[0]

file_path = os.path.join(PATH_TO_DATASET, filename)

df = pd.read_json(file_path)
```

```python
from sklearn.preprocessing import LabelEncoder



def get_encoder(df, player_columns:list[str]):
    encoder = LabelEncoder()    
    all_player_ids = pd.concat([df[col] for col in player_columns])
    encoder.fit(all_player_ids)
    return encoder

```

```python
player_columns = [f'player{i}_id' for i in range(1, 11)]

encoder = get_encoder(df, player_columns)
for col in player_columns:
    df[col] = encoder.transform(df[col])
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Embedding, Dense, Concatenate, Flatten
from tensorflow.keras.optimizers import Adam


def get_nn(player_columns: list[str], encoder: LabelEncoder):    
    embedding_dim = 32
    inputs = []
    embeddings = []

    for col in player_columns:
        input_layer = Input(shape=(1,), name=f"{col}_input")
        embed_layer = Embedding(input_dim=len(encoder.classes_), output_dim=embedding_dim, name=f"{col}_embed")(input_layer)
        embed_flat = Flatten()(embed_layer)
        inputs.append(input_layer)
        embeddings.append(embed_flat)
   
    merged = Concatenate()(embeddings)
    x = Dense(128, activation='relu')(merged)
    x = Dense(64, activation='relu')(x)
    output = Dense(1, activation='sigmoid')(x)

    model = Model(inputs=inputs, outputs=output)
    model.compile(optimizer=Adam(0.0001), loss='binary_crossentropy', metrics=['accuracy'])
    model.summary()
    return model
```

```python
nn = get_nn(player_columns, encoder)
```

```python
val_size = 100
train_df = df[:-val_size]
val_df = df[-val_size:]
```

```python
X_train = [train_df[col].values for col in player_columns]
y_train = train_df['target'].values

X_val = [val_df[col].values for col in player_columns]
y_val = val_df['target'].values

from tensorflow.keras.callbacks import EarlyStopping

# Define early stopping
early_stopping = EarlyStopping(
    monitor='val_loss',      # You can also monitor 'val_accuracy'
    patience=5,             # Stop after 10 epochs of no improvement
    restore_best_weights=True # Restore model from best epoch
)

# === Train with EarlyStopping ===
nn.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    epochs=100,
    batch_size=128,
    callbacks=[early_stopping],
    verbose=1
)
```
